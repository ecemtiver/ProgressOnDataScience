```python
#Değişkenler üzerinde işlemler(sütun bilgileri)

import pandas as pd
import seaborn as sns

pd.set_option('display.max_columns', None) #üç noktalar gitsin
df = sns.load_dataset("titanic")
df.head()
```

```python
"age" in df #herhangi bir değişkeni sorgulama

#df["age"].head()
#df.age.head()
```

```python
df["age"].head()
type(df["age"].head())
```

```python
#tek bir seçim yaparken, seçim sonunun df olarak kalması
df[["age"]].head()
type(df[["age"]].head())
```

```python
df[["age","alive"]]
```

```python
#birden fazla değişken seçerken
columns_names = ["survived", "sex", "alive"]
df[columns_names]
```

```python
#değişken ile yeni değişken ekleme

df["survived2"] = df["survived"]**3
df
df["survived3"] = df["survived2"]/df["survived"]
df
df.drop("survived2", axis=1).head()
```

```python
#liste, drop ile çoklu silme

df.drop(columns_names, axis=1).head()
```

```python
df.loc[df["survived"]>0, ["survived","class"]].head() #iki değişken
```

```python
#birden fazla koşul parantez ile ifade edilir

df.loc[(df["survived"]>0) & (df["sex"]=="female"), ["survived", "class"]].head()
```

```python
df.loc[(df["survived"]>0) 
       & (df["sex"]=="female") 
       & (df["embark_town"]=="Southampton"), 
       ["survived", "class", "embark_town"]].head()
```

```python
df.loc[(df["survived"]>0) 
       & (df["sex"]=="female") 
       & ((df["embark_town"]=="Southampton") | (df["embark_town"]=="Cherbourg")), 
       ["survived", "class", "embark_town"]].head()
```
