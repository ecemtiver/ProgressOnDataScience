```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("titanic")
df.index = df["age"]
df.drop("age", axis=1).head() #silme işlemi sütunda axis=1 şeklinde

#özet; yaş değişkeniyle index yer değişti ve yaş değişkeni silindi
```

```python
df.drop("age", axis=1, inplace=True) #kalıcı
df.index
```

```python
#indexi değişkene çevirme
#df.index seçip data frame'e değişken at


df["age"] = df.index #eğer girilen ifade df'de yoksa yeni değişken olarak eklenir
df.head()
```
