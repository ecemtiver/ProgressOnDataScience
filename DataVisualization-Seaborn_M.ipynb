```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
df = sns.load_dataset("tips")
df.head()
```

```python
df["sex"].value_counts()
sns.countplot(x=df["sex"], data=df) #seaborn
plt.show()
```

```python
df["sex"].value_counts().plot(kind="bar") #plotlib
plt.show()
```

```python
sns.boxplot(x=df["total_bill"]) #outlier
plt.show()
```

```python
df["total_bill"].hist()
plt.show()
```

```python
df["tip"].hist()
plt.show()
```

```python
sns.scatterplot(x=df["tip"], y=df["total_bill"], hue=df["smoker"], data=df)
plt.show()
```
