```python
#apply&lambda
 #apply; satır ya da sütunda otomatik fonksiyon çalıştırır
  #lamda; kullan at fonksiyon tanımlar(def gibi)


import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
df = sns.load_dataset("titanic")
df.head()
```

```python
df["age2"] = df["age"]*3
df["age3"] = df["age2"]*6
df.head()
```

```python
#Uzun yol
e = df["age"]/10
c = df["age2"]/10
m = (df["age3"]/10).head() #head kullanımı

print(e)
print(c)
print(m)
```

```python
#değişkenlere döngü uygulama
 #for ile
for col in df.columns:
    if "age" in col:
        print(col)
```

```python
for col in df.columns:
    if "age" in col:
        print((df[col]/10).head())
```

```python
for col in df.columns:
    if "age" in col:
        df[col] = df[col]/10 #atama işlemi var
df.head()
```

```python
#apply&lamda ile

df[["age", "age2", "age3"]].apply(lambda x: x/10).head()
```

```python
#loc ile

df.loc[:, df.columns.str.contains("age")].apply(lambda x: x/10).head()
```

```python
df.loc[:, df.columns.str.contains("survived")].apply(lambda x: x**2).head()
```

```python
#uygulandığın df'deki değerleri standartlaştırsın

df.loc[:, df.columns.str.contains("age")].apply(lambda x: (x -x.mean())/x.std()).head()
```

```python
#standart_scaler fonksiyonu atama

def standart_scaler(col_name):
    return (col_name - col_name.mean())/col_name.std()
df.loc[:, df.columns.str.contains("age")].apply(standart_scaler).head()
```

```python
#işlemi kaydetmek için 2 çeşit yol

#df.loc[:, ["age", "age2", "age3"]] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)

df.loc[:, df.columns.str.contains("age")] = df.loc[:, df.columns.str.contains("age")].apply(standart_scaler)
df.head()
```

```python
#join with concat

import numpy as np
import pandas as pd

j = np.random.randint(1,25, size=(5,3)) #numpy array
df1 = pd.DataFrame(j, columns=["var", "var2", "var3"]) 

#pd.DataFrame verilen veri yapısına ve değişkenlere göre sıfırdan df oluşturma
df2 = df1 + 25

pd.concat([df1,df2], ignore_index=True)
```

```python
#join with merge

df_1  = pd.DataFrame({'students': ['Ece', 'Mert', 'İrem', 'Ecem'], 'deparment': ['finance', 'marine', 'mathematics', 'data science']})
df_2 = pd.DataFrame({'students': ['Ece', 'Mert', 'İrem', 'Ecem'], 'graduation_year': [2022, 2021, 2023, 2022]})

pd.merge(df_1, df_2, on='students')
```

```python
df_3 = pd.merge(df_1,df_2)

df_4 = pd.DataFrame({'deparment': ['finance', 'marine', 'mathematics', 'data science'], 'school': ['deu', 'pau', 'bau', 'itu']})

pd.merge(df_3,df_4)
```
