```python
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
df = sns.load_dataset("titanic")
df.head()
```

```python
df["age"].mean()
```

```python
df.groupby("sex")["age"].mean() #cinsiyete göre grupla yaşa göre ort al
```

```python
df.groupby("sex").agg({"age": "mean"}) #agg içinde key ve value var; sözlük
```

```python
df.groupby("sex").agg({"age": ["mean", "sum"]}) #veriyi cinsiyete göre kırma
```

```python
df.groupby("sex").agg({"age": ["mean", "sum"],
                      "embark_town": "count"}) 

#embark_town değil cinsiyetin frekansı gelir. pivot table lazım(!) "to In [23]"
```

```python
df.groupby("sex").agg({"age": ["mean", "sum"],
                      "survived": "mean"})
```

```python
df.groupby(["sex", "embark_town"]).agg({"age": ["mean"],
                      "survived": "mean"}) #iki seviyeli grupby
```

```python
df.groupby(["sex", "embark_town", "pclass"]).agg({"age": ["mean"],
                      "survived": "mean"})
```

```python
df.groupby(["sex", "embark_town", "pclass", "class"]).agg({"age": ["mean"],
                      "survived": "mean"})
```

```python
df.groupby(["sex", "embark_town", "pclass"]).agg({"age": ["mean"],
                      "survived": "mean", "sex": "count"})
```

```python
#pivot table

df.pivot_table("survived", "sex", "embarked") 
#kesişim(ön tanımlı değer ort ama şekillendirilebilir),
 #satırda(index), 
  #sütunda görmek istenen
```

```python
df.pivot_table("survived", "sex", "embarked", aggfunc="std") 
#ön tanımlı değer olan mean'i standart sapma olarak değiştirme
```

```python
df.pivot_table("survived", "sex", ["embarked", "class"])
#satır ve sütunda iki seviyeli grupby
```

```python
#istenilen analize göre grupby'ları değiştirmeye devam edeceğim

#cut ve qcut eldeki sayısal değişkenleri, kategorik değişkene çevirmede en yaygın fonksiyonlar
#cut; sayısal değişkenleri hangi kategorilere göre bölmek istenildiği belirtilir
#qcut; sayısal değişken tanınmıyor, çeyreklik değerlere göre bölme
```

```python
df.head()
df["new_age"] = pd.cut(df["age"], [0, 10, 20, 30, 40, 80])
df.head()
```

```python
df.pivot_table("survived", "sex", "new_age") #yaş ve cinsiyet kılımında hayatta kalma
 #hatırlatma: neye göre, satır, sütun
```

```python
df.pivot_table("survived", "sex", ["new_age", "pclass"])
```
